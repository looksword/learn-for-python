```python
import matplotlib.pyplot as plt
import json
import pandas as pd
import datetime
import csv
```

```python
path = 'D://小羽飞刀/测试/功率计对比数据/Data_2'
suffix = '.csv'
testData = pd.DataFrame([[0,0,0,0]],columns=['Load','Speed','ZC','NJYMT'])
with open(path+suffix, 'r', newline='',encoding='utf-8') as csvfile:
    spamreader = csv.reader(csvfile)
    next(spamreader) # 跳过表头
    for row in spamreader:
        tempLoad = int(row[0])
        tempSpeed = int(row[1])
        tempZC = float(row[2])
        tempNJYMT = float(row[3])
        new = pd.DataFrame([[tempLoad,tempSpeed,tempZC,tempNJYMT]],columns=['Load','Speed','ZC','NJYMT'])
        testData = pd.concat([testData, new], ignore_index=True)
```

```python
print(testData['Load'])
```

```python
testData2 = testData.loc[testData['Load']<200].reset_index(drop=True)
```

```python
print(testData2)
```

```python
devno = 0
colors = ['r','g','b','y','k']
testplot = testData2
for dataname in testplot.columns.values:
    fig = plt.figure(devno + 1)
    fig.set_size_inches(200,40,forward=True)
    x = range(len(testplot))
    y = list(testplot[dataname])
    newline = plt.plot(x,y)
    plt.setp(newline,color = colors[devno])
    plt.title(dataname)
    plt.show()
    devno += 1
```
